# Run EoH with Groq

This notebook sets the Groq environment variables and runs `runEoH_groq.py` from this repo.

```python
from pathlib import Path
import os
import sys

def find_repo_root(start: Path) -> Path:
    for path in [start, *start.parents]:
        if (path / "eoh" / "setup.py").exists() and (path / "examples" / "bp_online" / "runEoH_groq.py").exists():
            return path
    raise RuntimeError("Could not find the repo root from the current notebook location.")

REPO_ROOT = find_repo_root(Path.cwd().resolve())
BP_DIR = REPO_ROOT / "examples" / "bp_online"
EOH_SRC = REPO_ROOT / "eoh" / "src"

if str(EOH_SRC) not in sys.path:
    sys.path.insert(0, str(EOH_SRC))

os.chdir(BP_DIR)
print(f"Repo root: {REPO_ROOT}")
print(f"Working directory: {Path.cwd()}")
```

```python
# Fill in your key here, or leave it blank if GROQ_API_KEY / EOH_API_KEY is already set in JupyterLab.
GROQ_API_KEY = ""

if GROQ_API_KEY:
    os.environ["GROQ_API_KEY"] = GROQ_API_KEY

os.environ.setdefault("EOH_API_ENDPOINT", "https://api.groq.com/openai/v1")
os.environ.setdefault("EOH_MODEL", "qwen/qwen3-32b")
os.environ.setdefault("EOH_POP_SIZE", "2")
os.environ.setdefault("EOH_N_POP", "2")
os.environ.setdefault("EOH_N_PROC", "1")
os.environ.setdefault("EOH_EVAL_TIMEOUT", "250")
os.environ.setdefault("EOH_DEBUG", "false")
os.environ.setdefault("EOH_OUTPUT_PATH", str(BP_DIR))
os.environ.setdefault("EOH_REASONING_EFFORT", "none")
os.environ.setdefault("EOH_REASONING_FORMAT", "hidden")
os.environ.setdefault("EOH_MAX_COMPLETION_TOKENS", "1200")

assert os.getenv("GROQ_API_KEY") or os.getenv("EOH_API_KEY"), "Set GROQ_API_KEY or EOH_API_KEY before running."

for name in [
    "EOH_API_ENDPOINT",
    "EOH_MODEL",
    "EOH_POP_SIZE",
    "EOH_N_POP",
    "EOH_N_PROC",
    "EOH_EVAL_TIMEOUT",
    "EOH_DEBUG",
    "EOH_OUTPUT_PATH",
    "EOH_REASONING_EFFORT",
    "EOH_REASONING_FORMAT",
    "EOH_MAX_COMPLETION_TOKENS",
]:
    print(f"{name}={os.environ[name]}")
```

```python
import runpy

runpy.run_path(str(BP_DIR / "runEoH_groq.py"), run_name="__main__")
```
